```python
def calculate_skewness_kurtosis(data, column_name):
    """
    Calculates and prints the skewness and kurtosis for a specified column in the given DataFrame.

    Parameters:
    - data: DataFrame containing the data.
    - column_name: String, the name of the column for which skewness and kurtosis are to be calculated.

    Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean. 
    Positive skewness indicates a distribution with an asymmetric tail extending toward more positive values.
    
    Kurtosis measures the "tailedness" of the probability distribution. A higher kurtosis indicates a distribution with heavier tails, 
    and a lower kurtosis indicates a distribution with lighter tails.
    """
    skewness = data[column_name].skew()
    kurtosis = data[column_name].kurtosis()

    print(f"Skewness for {column_name}:", skewness)
    print(f"Kurtosis for {column_name}:", kurtosis)
```

```python
calculate_skewness_kurtosis(data,'Ladder score')
```

```python
def plot_correlation_heatmap(data):
    # Selecting only the numerical columns
    numerical_data = data.select_dtypes(include=['float64', 'int64'])
    
    # Calculating the correlation matrix
    corr_matrix = numerical_data.corr()
    
    # Generating the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
    return
```

```python
plot_correlation_heatmap(data)
```

# Relational Graph Function

```python
def scatter_plot(data):
    # Extracting the relevant columns
    x = data['Explained by: Healthy life expectancy']
    y = data['Ladder score']

    # Creating the scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, alpha=0.7, color='blue')

    # Adding labels and title
    plt.title('Scatter Plot: Ladder Score vs Explained by: Healthy Life Expectancy')
    plt.xlabel('Explained by: Healthy life expectancy')
    plt.ylabel('Ladder Score')

    # Adding grid for better readability
    plt.grid(True)

    # Displaying the plot
    plt.show()
    return
```

#           Categorical Graph Function

```python
#creating a bar plot of ladder scores within different countries:

def generate_bar_plot_top_countries(data, feature='Ladder score', top_n=10):
    top_countries = data.groupby('Country name')['Ladder score'].mean().nlargest(top_n).index
    top_countries_data = data[data['Country name'].isin(top_countries)]
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x=feature, y='Country name', data=top_countries_data, palette='viridis')

    plt.title(f'Distribution of {feature} for Top {top_n} Countries')
    plt.xlabel(feature)
    plt.ylabel('Country')

    plt.show()
    return
```

```python
generate_bar_plot_top_countries(data, feature='Ladder score', top_n=10)
```

```python
def generate_pie_chart_top_countries(data, top_n=10):
    # Selecting the top N countries based on ladder score mean
    top_countries = data.groupby('Country name')['Ladder score'].mean().nlargest(top_n).index
    top_countries_data = data[data['Country name'].isin(top_countries)]
    
    # Calculating the scores for the top N countries
    top_countries_scores = top_countries_data.groupby('Country name')['Ladder score'].sum()

    # Plotting the pie chart without 'Other' category
    plt.figure(figsize=(8, 8))
    plt.pie(top_countries_scores, labels=top_countries_scores.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Top {top_n} Countries')

    plt.show()
    return
```

```python
generate_pie_chart_top_countries(data, top_n=10)
```

# Elbow Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv('WHR2024.csv')  # Replace 'your_dataset.csv' with the path to your dataset

# Drop rows with missing values
data.dropna(inplace=True)

# Selecting the features for clustering
X = data.iloc[:, 1:].values  # Exclude the 'Country name' column

# Perform K-means clustering with a varying number of clusters
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plotting the inertia values against the number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)

# Finding the optimal number of clusters (elbow point)
delta_inertia = np.diff(inertia, 2)
optimal_k = np.argmin(delta_inertia) + 2  # Adding 2 because of the double difference
plt.axvline(x=optimal_k, color='r', linestyle='-')
plt.text(optimal_k + 0.1, inertia[optimal_k - 1], 'Optimal k = {}'.format(optimal_k), rotation=90)

plt.show()

print("Optimal number of clusters (k):", optimal_k)
```

# K-Mean Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv('WHR2024.csv')  # Replace 'your_dataset.csv' with the path to your dataset

# Drop rows with missing values
data.dropna(inplace=True)

# Selecting the features for clustering
X = data.iloc[:, 3:5].values  # Selecting two features for simplicity (you can choose other features)

# Optimal number of clusters obtained from the elbow method
optimal_k =6 # Replace this with the value obtained from the elbow method

# Perform K-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Add cluster labels to the dataset
data['Cluster'] = cluster_labels

# Visualize the clusters
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data.iloc[:, 3], cluster_data.iloc[:, 4], label='Cluster {}'.format(cluster + 1))

plt.title('Clustering of Countries')
plt.xlabel('Explained by: Log GDP per capita')
plt.ylabel('Explained by: Social support')
plt.legend()
plt.grid(True)
plt.show()
```

# Fitting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset
data = pd.read_csv('WHR2024.csv')  # Replace 'your_dataset.csv' with the path to your dataset

# Drop rows with missing values
data.dropna(inplace=True)

# Selecting the features for clustering
X = data.iloc[:, 1:].values  # Exclude the 'Country name' column

# Optimal number of clusters obtained from the elbow method
optimal_k = 6  # Replace this with the value obtained from the elbow method

# Perform K-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Add cluster labels to the dataset
data['Cluster'] = cluster_labels

# Calculate silhouette score
silhouette_avg = silhouette_score(X, cluster_labels)

# Print the silhouette score
print("Silhouette Score:", silhouette_avg)

# Get centroids of each cluster
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['Ladder score'], cluster_data['Explained by: Log GDP per capita'], label='Cluster {}'.format(cluster + 1))

# Plot centroids as points
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', label='Centroids')

plt.title('K-means Clustering of Countries with Centroids')
plt.xlabel('Ladder score')
plt.ylabel('Explained by: Log GDP per capita')
plt.legend()
plt.grid(True)
plt.show()
```
